# which Grade has Highest Students?

```python
sns.countplot(data = df, x = "grade")
plt.title("grade distribution")
plt.show()
```

# Study Hours vs Total Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data = df, x = "weekly_self_study_hours", y = "total_score")
plt.title("study hours VS total score")
plt.show()
```

# Class Participation VS Total Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data = df , x = "class_participation" , y = "total_score")
plt.title("Participation VS Score")
plt.show()
```

# Participation VS Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data = df , x = "attendance_percentage" , y = "total_score")
plt.title("Participation VS Score")
plt.show()
```

# Correlation

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
